```python
import requests
import pandas as pd
```

```python

url = 'http://api.citybik.es/v2/networks/mobibikes' 
response = requests.get(url)
data = response.json()

# Extract stations
stations = data['network']['stations']
stations_df = pd.DataFrame(stations)

# Clean and select relevant columns
stations_df = stations_df[['id', 'name', 'latitude', 'longitude', 'free_bikes', 'empty_slots']]

# Optional: restrict to first 5 stations for testing
#stations_subset = stations_df.head(5)
#stations_subset
```
